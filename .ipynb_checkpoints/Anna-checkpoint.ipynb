### Moran's I
It assesses the similarity or dissimilarity of attribute values between neighboring locations. 
(Try with different k)

```python
w = weights.KNN.from_dataframe(gdf,k=7)
# Calculate Moran's I for each variable
mi_results = [esda.Moran(gdf[variable], w) for variable in cluster_cols]
# Structure results as a list of tuples
mi_results = [(variable, res.I, res.p_sim) for variable,res in zip(cluster_cols, mi_results)]
# Display on table
table = pd.DataFrame(
    mi_results,
    columns=['Variable', "Moran's I", 'P-value']
).set_index('Variable')
table
```

## K-means

```python
elbow_df = gdf[['TFR', 'master_perc', 'disp_income_avg_k', 'dwelling_above_74_perc', 'household_wealth_avg_k']]
scaled_elbow_df = robust_scale(elbow_df)
```

Find optimal value of k

```python
#Define a range of values for k
k_values = range(1, 11)
inertias = []

# Perform k-means clustering for each value of k
for k in k_values:
    kmeans = cluster.KMeans(n_clusters=k)
    kmeans.fit(elbow_df)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans4 = cluster.KMeans(n_clusters=4, random_state=12345)
```

```python
# scale data
for c in cluster_cols:
    gdf[c+'_scaled'] = robust_scale(gdf[c])
```

```python
cluster_cols_scaled = [c+'_scaled' for c in cluster_cols]
```

```python
# Run the clustering algorithm
k4cls = kmeans4.fit(gdf[cluster_cols_scaled])
```

```python
k4cls.labels_
```

```python
# append to original data
gdf['k4cls'] = k4cls.labels_
```

```python
# Display values on map

f, ax = plt.subplots(1, figsize=(9, 9))

gdf.plot(
    column='k4cls', 
    categorical=True, 
    legend=True, 
    linewidth=0.2, 
    ax=ax, 
    cmap='tab20b'
)
# Remove axis
ax.set_axis_off()
# Add title
plt.title('Need to find a title')

# Add basemap
cx.add_basemap(
    ax,
    crs=gdf.crs,
    source=cx.providers.CartoDB.DarkMatterNoLabels # can change this to something (white if thats better)
)

# Display the map
plt.show()
```

Several spatially connected clusters appear.

```python
# Calculate the mean by group
k4means = gdf.groupby('k4cls')[cluster_cols].mean()

k4means.T
```

```python
# Index db on cluster ID
tidy_db = gdf.set_index('k4cls')
# Keep only variables used for clustering
tidy_db = tidy_db[cluster_cols]
# Stack column names into a column, obtaining
# a "long" version of the dataset
tidy_db = tidy_db.stack()
# Take indices into proper columns
tidy_db = tidy_db.reset_index()
# Rename column names
tidy_db = tidy_db.rename(
    columns={"level_1": "Attribute", 0: "Values"}
)
# Check out result
tidy_db.head(30)


# Scale fonts to make them more readable
sns.set(font_scale=1.5)
# Setup the facets
facets = sns.FacetGrid(
    data=tidy_db,
    col="Attribute",
    hue='k4cls',
    sharey=False,
    sharex=False,
    aspect=2,
    col_wrap=3,
)
# Build the plot from `sns.kdeplot`
facets.map(sns.kdeplot, "Values", fill=True,warn_singular=False).add_legend();
```
